```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('./trained_models/training_process.pkl', 'rb') as f:
    training_info = pickle.load(f)

epochs = range(1, len(training_info['train_loss_hist']) + 1)

# Plot Loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, training_info['train_loss_hist'], label='Train Loss')
plt.plot(epochs, training_info['val_loss_hist'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, training_info['train_acc_hist'], label='Train Accuracy')
plt.plot(epochs, training_info['val_acc_hist'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend()

plt.grid(True)
plt.show()
```
